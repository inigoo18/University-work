# Part 1: Kernels

Below we implement a couple of kernels

```python
def exponentiated_quadratic(xa, xb, sigma=1, length_scale=1):
    """Exponentiated quadratic  with Ïƒ=1"""
    # L2 distance (Squared Euclidian)
    sq_norm = -0.5 * scipy.spatial.distance.cdist(xa, xb, 'sqeuclidean')/length_scale
    return (sigma**2) * np.exp(sq_norm)
```

```python
def periodic_kernel(xa, xb, period=1, length_scale=1, sigma=1):
    # L1 norm
    l1_norm =  scipy.spatial.distance.cdist(xa, xb, 'minkowski', p=1)
    return sigma**2 * np.exp(-2/length_scale**2 * (np.sin(np.pi * l1_norm/period)**2))
```

```python
def linear_kernel(xa, xb):
    return xa@xb.T
```

Let's make a few illustrations following the same theme as in the slides

```python
dim = 50
num_samples = 5
X = np.expand_dims(np.linspace(-4, 4, dim), 1) # -4, -3.8, -3.6, ...., 4
```

```python
COV = periodic_kernel(X, X, length_scale=1, period=5) # covariance between samples: 50 dimensions
plt.figure()
sns.heatmap(COV, cmap='Blues')
```

```python
mean = np.zeros((COV.shape[0],)) # mean funct: array of 0s of len 50
x = np.random.multivariate_normal(mean.squeeze(), COV, size=num_samples) # 5 samples of dim 50 each; COV : std funct.
for i in range(num_samples):
    plt.plot(X.squeeze(), x[i], '-')
plt.xlabel(r'$x$', fontsize=13)
plt.ylabel(r'$f(x)$', fontsize=13)
```

```python
plt.plot(X, COV[25, :], '-', label=f'$l=${1}') # halfway in the linear space -4 -> +4
plt.title('Exponentiated quadratic distance plot')
plt.xlabel(r'$x_a-x_b$')
plt.ylabel(r'$k(x_a,x_b)$')
```

### Part 1.1

From the three kernels that we have studied:
- exponentiated-quadratic has a length scale parameter, and the greater the value, the smoother the covariance matrix (e.g: dim 0 has some correlation with dim 18 while with a low length scale it wouldn't.)

- periodic kernel has a length scale parameter and a period parameter.
- - length scale parameter: similar to the exponentiated-quadratic kernel, the greater the value, the more ambiguity between the dimensions in the covariance matrix.
- - period parameter: specify the duration of the periodic patterns. It represents the distance at which the function completes one cycle. A shorter period corresponds to more repetitions



### Part 1.2

Periodic + linear kernel: we see signs of the periodic kernel with a smoothed value all in between (periodic + trend) Adding them together gets characteristics of both kernels.

Periodic x linear kernel: growing amplitude

```python
def periodic_linear_kernel_add(xa, xb, period=1, length_scale=1, sigma=1):
    return periodic_kernel(xa,xb,period,length_scale,sigma) + linear_kernel(xa,xb)

def periodic_linear_kernel_mul(xa, xb, period=1, length_scale=1, sigma=1):
    return periodic_kernel(xa,xb,period,length_scale,sigma) * linear_kernel(xa,xb)
```

```python
COV = periodic_linear_kernel_add(X, X, length_scale=.5, period=1) # covariance between samples: 50 dimensions
plt.figure()
sns.heatmap(COV, cmap='Blues')
```

```python
COV = periodic_linear_kernel_mul(X, X, length_scale=.5, period=1) # covariance between samples: 50 dimensions
plt.figure()
sns.heatmap(COV, cmap='Blues')
```

Exponentiated + linear kernel: linear with added variation.

Exponentiated x linear kernel: very restrictive, growing variation.

```python
def exponentiated_linear_kernel_add(xa, xb, length_scale=1, sigma=1):
    return exponentiated_quadratic(xa,xb,length_scale,sigma) + linear_kernel(xa,xb)

def exponentiated_linear_kernel_mul(xa, xb, period=1, length_scale=1, sigma=1):
    return exponentiated_quadratic(xa,xb,length_scale,sigma) * linear_kernel(xa,xb)
```

```python
COV = exponentiated_linear_kernel_add(X, X, length_scale=1) # covariance between samples: 50 dimensions
plt.figure()
sns.heatmap(COV, cmap='Blues')
```

```python
COV = exponentiated_linear_kernel_mul(X, X, length_scale=1) # covariance between samples: 50 dimensions
plt.figure()
sns.heatmap(COV, cmap='Blues')
```

Exponentiated + periodic kernel: periodic with increasing variation

Exponentiated x periodic kernel: periodic with barely any variation

```python
def exponentiated_periodic_kernel_add(xa, xb, length_scale=1, period=1, sigma=1):
    return exponentiated_quadratic(xa,xb,length_scale,sigma) + periodic_kernel(xa, xb, period, length_scale, sigma)

def exponentiated_periodic_kernel_mul(xa, xb, length_scale=1, period=1, sigma=1):
    return exponentiated_quadratic(xa,xb,length_scale,sigma) * periodic_kernel(xa, xb, period, length_scale, sigma)
```

```python
COV = exponentiated_periodic_kernel_add(X, X, length_scale=1, period = 1) # covariance between samples: 50 dimensions
plt.figure()
sns.heatmap(COV, cmap='Blues')
```

```python
COV = exponentiated_periodic_kernel_mul(X, X, length_scale=1) # covariance between samples: 50 dimensions
plt.figure()
sns.heatmap(COV, cmap='Blues')
```

## Part 2: Self study

In this exercise we will be experimenting with data on atmosoheric CO2 measurements data from Mauna Loa. You can find additional information about the data from the <a href="https://scrippsco2.ucsd.edu/data/atmospheric_co2/primary_mlo_co2_record.html">here</a>, included the full data set. In this exercise we will be using a slightly cleaned up version of the original dataset.

```python
co2_df = pd.read_csv(
    # Source: https://scrippsco2.ucsd.edu/assets/data/atmospheric/stations/in_situ_co2/monthly/monthly_in_situ_co2_mlo.csv
    './co2.csv',
    header=0, # Data starts here
    usecols=[3, 4], # Only keep the 'Date' and 'CO2' columns
    na_values='-99.99',  # NaNs are denoted as '-99.99'
    dtype=np.float64,
    sep=','
)
```

```python
# Drop missing values
co2_df.dropna(inplace=True)
```

```python
co2_df
```

```python
# Split the data into an observed part and a part we want to make predictions for
date_split_predict = 2010
df_observed = co2_df[co2_df.Date < date_split_predict]
print('{} measurements in the observed set'.format(len(df_observed)))
df_predict = co2_df[co2_df.Date >= date_split_predict]
print('{} measurements in the test set'.format(len(df_predict)))
```

```python
plt.figure()
plt.plot(df_observed.iloc[:,0].values, df_observed.iloc[:,1].values)
plt.title("In situ air measurements at Mauna Loa, Observatory, Hawaii")
```

Separate out the input data and the observations

```python
x_train = df_observed['Date'].values
f_train = df_observed['CO2'].values
x_test = df_predict['Date'].values
f_test = df_predict['CO2'].values
```

Setup a first shot at a kernel. Scikit learn offers a variety of predifed kernels. An overview of the available kernels can be found <a href="https://scikit-learn.org/stable/modules/gaussian_process.html#kernels-for-gaussian-processes">here</a> and a gentle introduction to Gaussian processes in sklearn can be found <a href="https://scikit-learn.org/stable/modules/gaussian_process.html#">here</a>. In particular, it would be a good idea to consult the kernel section to get a better understanding of the kernel specification below.

The kernel is composed of several terms that are responsible for explaining different properties of the signal:

- a **long term, smooth rising trend is to be explained by an RBF kernel**. The RBF kernel with a large length-scale enforces this component to be smooth; it is not enforced that the trend is rising which leaves this choice to the GP. The specific length-scale and the amplitude are free hyperparameters.

- a **seasonal component, which is to be explained by the periodic ExpSineSquared kernel** with a fixed periodicity of 1 year. The length-scale of this periodic component, controlling its smoothness, is a free parameter. In order to allow decaying away from exact periodicity, the product with an RBF kernel is taken. The length-scale of this RBF component controls the decay time and is a further free parameter.

- **smaller, medium term irregularities are to be explained by a RationalQuadratic kernel component**, whose length-scale and alpha parameter, which determines the diffuseness of the length-scales, are to be determined. According to [RW2006], these irregularities can better be explained by a RationalQuadratic than an RBF kernel component, probably because it can accommodate several length-scales. (NOT USED)

- **a â€œnoiseâ€ term, consisting of an RBF kernel contribution**, which shall explain the correlated noise components such as local weather phenomena, and a WhiteKernel contribution for the white noise. The relative amplitudes and the RBFâ€™s length scale are further free parameters.

```python
co2_kernel_old = 50**2*RBF(length_scale=20) + WhiteKernel(noise_level=10, noise_level_bounds="fixed")
co2_kernel = 60**2*RBF(length_scale=20) + 1**2*(RBF(length_scale=2)+ ExpSineSquared(length_scale=5, periodicity=1)*(RBF(length_scale=2))) + (0.5**2*WhiteKernel(noise_level=10, noise_level_bounds="fixed")*(RBF(length_scale=2)))
```

With the kernel specified, we can start a Gaussian process model of our training data. 

```python
f_mean = f_train.mean()
gaussian_process = GaussianProcessRegressor(kernel=co2_kernel, normalize_y=False)
gaussian_process.fit(x_train.reshape(-1,1), f_train - f_mean)
```

Using our newly learned GP, we can do prediction on the test data.

```python
mean_f_pred, std_f_pred = gaussian_process.predict(x_test.reshape(-1,1), return_std=True)
mean_f_pred += f_mean
```

Let's get a feel for the results

```python
plt.plot(x_train, f_train, '-b')
plt.plot(x_test, f_test, color="tab:blue", alpha=0.4)
plt.plot(x_test, mean_f_pred, '-r')
plt.fill_between(
    x_test.ravel(),
    mean_f_pred - std_f_pred,
    mean_f_pred + std_f_pred,
    color="tab:red",
    alpha=0.2,
)
plt.xlabel("Year")
plt.ylabel("CO$_2$ concentration (ppm)")
_ = plt.title(
    "Monthly average of air samples measurements\nfrom the Mauna Loa Observatory")
plt.show()
```

```python
print(f"RMSE: {np.mean((mean_f_pred - f_test)**2)}")
```
