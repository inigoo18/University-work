```python
import torch
from torchvision import datasets, transforms
from matplotlib import pyplot
import matplotlib.pyplot as plt
import numpy as np
import time
from sklearn.metrics import confusion_matrix,accuracy_score
```

Through torch load the MNIST data set, which we will use in this self study. The MNIST database consists of grey scale images of handwritten digits. Each image is of size $28\times 28$; see figure below for an illustration. The data set is divided into a training set consisting of $60000$ images and a test set with $10000$ images; in both

Using the data loader provided by torch we have an easy way of loading in data in batches (here of size 64). We can also make various other transformation of the data, such as normalization. The details for loading the data are not really that important, but if yiu are curious, you can find a quick introduction here.

```python
batch_size = 64
train_loader = torch.utils.data.DataLoader(
    datasets.MNIST('../data', train=True, download=True,
    transform=transforms.Compose([
    transforms.ToTensor(), # transforms to value [0 ,1]
    transforms.Normalize((0.1307,), (0.3081,)) # normalize using mean & std (substract mean and divide std)
    ])),
    batch_size=batch_size, shuffle=True)
test_loader = torch.utils.data.DataLoader(
    datasets.MNIST('../data', train=False, transform=transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.1307,), (0.3081,))
    ])),
    batch_size=batch_size, shuffle=True)
```

In the cell above, we are loading data in batches, which is subsequently used during optimization. We didn't discuss batch based learning during the last lecture, but please take a look at Slides 24 and 25 to get the gist of the approach. We will cover it in more detail during the next lecture. 

Each batch is a list of two elements. The first element encodes the digit and has dimensions [64,1,28,28] (the figures are greyscale with no rbg channel, hence the '1'), and the second element contains the class/label information. 

Meaning batch[0] is for the image data, and batch[1] for the labels.

```python
batch = next(iter(train_loader))
print(f"Batch dimension (digit): {batch[0].shape}")
print(f"Batch dimension (target): {batch[1].shape}")
digit_batch = batch[0]
img = digit_batch[0,:]
pyplot.imshow(img.reshape((28, 28)), cmap="gray")
print(f"Target: {batch[1][0]} with shape {batch[0][0].shape}")
```

With PyTorch we can specify that the tensors require gradients. This will make PyTorch record all operations performed on the tensors, so that we can afterwards calculate the gradients automatically using back propagation. See also the code example from the last lecture.

For the first part of this self study we will specify a neural network, which will encode a softmax function (see, e.g., Page 180 in The Deep Learning book). For this we need a (randomly initialized) weight matrix and a bias, and for both of them we need their gradients wrt. our error function (yet to be defined) in order to perform learning. Note that to facilitate matrix multiplication we will flatten our image from $28\times 28$ to $784$. 

```python
weights = torch.randn(784, 10) / np.sqrt(784) # 28 x 28 = 784; 10 because ten classes. weights[0..784] with 10 values each
# We divide by sqrt(784) to initialize weights to small values so that they are within reasonable range.
weights.requires_grad_() # set requirement for Pytorch to record operations ==> backpropagation.
bias = torch.zeros(10, requires_grad=True) # bias for each of the classes
```

```python
weights
```

Out model specification

```python
def softmax(x):
    return x.exp() / x.exp().sum(-1).unsqueeze(-1)

def model(xb):
    return softmax(xb @ weights + bias)
```

Let's test our model (with our randomly initialized weights)

```python
# We flatten the digit representation so that it is consistent with the weight matrix
xb = digit_batch.flatten(start_dim=1) # flattening operation starts from dimension 1. (start_dim = 0 would return a list.)
print(f"Batch shape: {xb.shape}")
preds = model(xb)
print(f"Prediction on first image {preds[0]}")
print(f"Corresponding classification: {preds[0].argmax()}")
```

Next we define our loss function, in this case the log-loss (or negative log-likelihood):

```python
def nll(input, target):
    return (-input[range(target.shape[0]), target].log()).mean() 
    # this amazing function looks at preds, which is 64 x 10, and looks in each row the output value for the corresponding
    # target. These are the probabilities that the actual label comes up, which we then take the log of and then average them.
loss_func = nll

# Make a test calculation
yb = batch[1]
print(loss_func(preds,yb)) # predictions compared to targets
```

In the end, we are interested in the accuracy of our model

```python
def accuracy(out, yb):
    preds = torch.argmax(out, dim=1)
    return (preds == yb).float().mean()
```

```python
print(f"Accuracy of model on batch (with random weights): {accuracy(preds, yb)}")
```

Now we are ready to combine it all and perform learning

```python
epochs = 1 # how many epochs to train for
lr = 0.01 # learning rate

train_losses = []
for epoch in range(epochs):
    for batch_idx, (xb, yb) in enumerate(train_loader):

        xb = xb.squeeze().flatten(start_dim=1)
        pred = model(xb)
        # We specify the loss
        loss = loss_func(pred, yb)
        # and perform backpropagation 
        loss.backward()
        # Lastly we update the weights and bias (torch.no_grad() ensures that no gradient 
        # calculations are taking place in this part of the code)
        with torch.no_grad():
            weights -= weights.grad * lr
            bias -= bias.grad * lr

            # After updating we set the gradients to zero so that we ar eready for the next round
            weights.grad.zero_()
            bias.grad.zero_()

            if batch_idx % 50 == 0:
                with torch.no_grad():
                    train_loss = np.mean([loss_func(model(txb.squeeze().flatten(start_dim=1)), tyb).item() for txb, tyb in train_loader])
                    print(f"Epoch: {epoch}, B-idx: {batch_idx}, Training loss: {train_loss}")
                    train_losses.append(train_loss)
```

Plot the evolution of the training loss

```python
plt.plot(range(len(train_losses)), train_losses)
```

```python
def softmax(x):
    return x.exp() / x.exp().sum(-1).unsqueeze(-1)

def nll(input, target):
    return (-input[range(target.shape[0]), target].log()).mean() 
    # this amazing function looks at preds, which is 64 x 10, and looks in each row the output value for the corresponding
    # target. These are the probabilities that the actual label comes up, which we then take the log of and then average them.
loss_func = nll
```

```python
def initModel(batch_size):
    train_loader = torch.utils.data.DataLoader(
        datasets.MNIST('../data', train=True, download=True,
        transform=transforms.Compose([
        transforms.ToTensor(), # transforms to value [0 ,1]
        transforms.Normalize((0.1307,), (0.3081,)) # normalize using mean & std (substract mean and divide std)
        ])),
        batch_size=batch_size, shuffle=True)
    test_loader = torch.utils.data.DataLoader(
        datasets.MNIST('../data', train=False, transform=transforms.Compose([
        transforms.ToTensor(),
        transforms.Normalize((0.1307,), (0.3081,))
        ])),
        batch_size=batch_size, shuffle=True)
    weights = torch.randn(784, 10) / np.sqrt(784) # 28 x 28 = 784; 10 because ten classes. weights[0..784] with 10 values each
    # We divide by sqrt(784) to initialize weights to small values so that they are within reasonable range.
    weights.requires_grad_() # set requirement for Pytorch to record operations ==> backpropagation.
    bias = torch.zeros(10, requires_grad=True) # bias for each of the classes
    
    return train_loader, weights, bias
```

```python
def model_new(xb, weights, bias):
    return softmax(xb @ weights + bias)
```

```python
def trainModel(epochs, lr, batch_size, loss_func, log = False):
    train_loader, weights, bias = initModel(batch_size)

    train_losses = []
    time_s = time.time()
    for epoch in range(epochs):
        for batch_idx, (xb, yb) in enumerate(train_loader):

            xb = xb.squeeze().flatten(start_dim=1)
            pred = model_new(xb, weights, bias)
            # We specify the loss
            loss = loss_func(pred, yb)
            # and perform backpropagation 
            loss.backward()
            # Lastly we update the weights and bias (torch.no_grad() ensures that no gradient 
            # calculations are taking place in this part of the code)
            with torch.no_grad():
                weights -= weights.grad * lr
                bias -= bias.grad * lr

                # After updating we set the gradients to zero so that we ar eready for the next round
                weights.grad.zero_()
                bias.grad.zero_()

                if batch_idx % 50 == 0:
                    with torch.no_grad():
                        train_loss = np.mean([loss_func(model_new(txb.squeeze().flatten(start_dim=1), weights, bias), tyb).item() for txb, tyb in train_loader])
                        if log:
                            print(f"Epoch: {epoch}, B-idx: {batch_idx}, Training loss: {train_loss}")
                        train_losses.append(train_loss)
    time_e = time.time()
    label = "batch "+str(batch_size)+", lr" + str(lr) + ", time: " + str(round((time_e - time_s), 2))
    return [train_losses, label]
```

```python
models = [
    trainModel(1,.01,512, nll),
    trainModel(1,.001,512, nll),
    trainModel(1, .01, 256, nll),
    trainModel(1, .001, 256, nll),
    trainModel(1, .01, 124, nll),
    trainModel(1, .001, 124, nll)
]
```

```python
def trainModel_momentum(epochs, lr, batch_size, loss_func, initModel, mom_alpha = .1, log = False):
    train_loader, weights, bias = initModel(batch_size)
    momentum = 0
    train_losses = []
    time_s = time.time()
    
    # Initialize the momentum for weights and bias
    weights_momentum = torch.zeros_like(weights)
    bias_momentum = torch.zeros_like(bias)
    
    for epoch in range(epochs):
        for batch_idx, (xb, yb) in enumerate(train_loader):

            xb = xb.squeeze().flatten(start_dim=1)
            pred = model_new(xb, weights, bias)
            # We specify the loss
            loss = loss_func(pred, yb)
            # and perform backpropagation 
            loss.backward()
            # Lastly we update the weights and bias (torch.no_grad() ensures that no gradient 
            # calculations are taking place in this part of the code)
            with torch.no_grad():
                # v <- alpha * v - lr * grad(loss)
                # w <- w + v
                weights_momentum = mom_alpha * weights_momentum - lr * weights.grad
                bias_momentum = mom_alpha * bias_momentum - lr * bias.grad
                
                weights += weights_momentum
                bias += bias_momentum

                # After updating we set the gradients to zero so that we ar eready for the next round
                weights.grad.zero_()
                bias.grad.zero_()

                if batch_idx % 50 == 0:
                    with torch.no_grad():
                        train_loss = np.mean([loss_func(model_new(txb.squeeze().flatten(start_dim=1), weights, bias), tyb).item() for txb, tyb in train_loader])
                        if log:
                            print(f"Epoch: {epoch}, B-idx: {batch_idx}, Training loss: {train_loss}")
                        train_losses.append(train_loss)
    time_e = time.time()
    label = "batch "+str(batch_size)+", lr" + str(lr) + ", time: " + str(round((time_e - time_s), 2))
    return [train_losses, label]
```

```python
models = [
    trainModel_momentum(1,.01,512, nll, initModel, .1),
    trainModel_momentum(1,.001,512, nll, initModel, .1),
    trainModel_momentum(1, .01, 256, nll, initModel, .1),
    trainModel_momentum(1, .001, 256, nll, initModel, .1),
    trainModel_momentum(1, .01, 124, nll, initModel, .1),
    trainModel_momentum(1, .001, 124, nll, initModel, .1)
]
```

```python
def initModel_higherparams(batch_size):
    train_loader = torch.utils.data.DataLoader(
        datasets.MNIST('../data', train=True, download=True,
        transform=transforms.Compose([
        transforms.ToTensor(), # transforms to value [0 ,1]
        transforms.Normalize((0.1307,), (0.3081,)) # normalize using mean & std (substract mean and divide std)
        ])),
        batch_size=batch_size, shuffle=True)
    test_loader = torch.utils.data.DataLoader(
        datasets.MNIST('../data', train=False, transform=transforms.Compose([
        transforms.ToTensor(),
        transforms.Normalize((0.1307,), (0.3081,))
        ])),
        batch_size=batch_size, shuffle=True)
    weights = torch.randn(784, 10) / 4 # 28 x 28 = 784; 10 because ten classes. weights[0..784] with 10 values each
    # We divide by sqrt(784) to initialize weights to small values so that they are within reasonable range.
    weights.requires_grad_() # set requirement for Pytorch to record operations ==> backpropagation.
    bias = torch.zeros(10, requires_grad=True) # bias for each of the classes
    
    return train_loader, weights, bias
```

```python
def initModelWithTest(batch_size):
    train_loader = torch.utils.data.DataLoader(
        datasets.MNIST('../data', train=True, download=True,
        transform=transforms.Compose([
        transforms.ToTensor(), # transforms to value [0 ,1]
        transforms.Normalize((0.1307,), (0.3081,)) # normalize using mean & std (substract mean and divide std)
        ])),
        batch_size=batch_size, shuffle=True)
    test_loader = torch.utils.data.DataLoader(
        datasets.MNIST('../data', train=False, transform=transforms.Compose([
        transforms.ToTensor(),
        transforms.Normalize((0.1307,), (0.3081,))
        ])),
        batch_size=batch_size, shuffle=True)
    weights = torch.randn(784, 10) / np.sqrt(784) # 28 x 28 = 784; 10 because ten classes. weights[0..784] with 10 values each
    # We divide by sqrt(784) to initialize weights to small values so that they are within reasonable range.
    weights.requires_grad_() # set requirement for Pytorch to record operations ==> backpropagation.
    bias = torch.zeros(10, requires_grad=True) # bias for each of the classes
    
    return train_loader, test_loader, weights, bias
```

```python
def trainModel_momentum_withTest(epochs, lr, batch_size, loss_func, initModel, mom_alpha = .1, log = False):
    train_loader, test_loader, weights, bias = initModelWithTest(batch_size)
    momentum = 0
    train_losses = []
    time_s = time.time()
    
    # Initialize the momentum for weights and bias
    weights_momentum = torch.zeros_like(weights)
    bias_momentum = torch.zeros_like(bias)
    
    for epoch in range(epochs):
        labelsTrain = []
        trueLabelsTrain = []
        for batch_idx, (xb, yb) in enumerate(train_loader):

            xb = xb.squeeze().flatten(start_dim=1)
            pred = model_new(xb, weights, bias)
            trueLabelsTrain += yb.tolist()
            for i in pred:
                labelsTrain += [int(torch.argmax(i))]

            # We specify the loss
            loss = loss_func(pred, yb)
            # and perform backpropagation 
            loss.backward()
            # Lastly we update the weights and bias (torch.no_grad() ensures that no gradient 
            # calculations are taking place in this part of the code)
            with torch.no_grad():
                weights_momentum = mom_alpha * weights_momentum + lr * weights.grad
                bias_momentum = mom_alpha * bias_momentum + lr * bias.grad
                
                weights -= weights_momentum
                bias -= bias_momentum

                # After updating we set the gradients to zero so that we ar eready for the next round
                weights.grad.zero_()
                bias.grad.zero_()

                if batch_idx % 50 == 0:
                    with torch.no_grad():
                        train_loss = np.mean([loss_func(model_new(txb.squeeze().flatten(start_dim=1), weights, bias), tyb).item() for txb, tyb in train_loader])
                        if log:
                            print(f"Epoch: {epoch}, B-idx: {batch_idx}, Training loss: {train_loss}")
                        train_losses.append(train_loss)
        labelsTest = []
        trueLabelsTest = []
        for batch_idx, (xb, yb) in enumerate(test_loader):
            xb = xb.squeeze().flatten(start_dim=1)
            pred = model_new(xb, weights, bias)
            trueLabelsTest += yb.tolist()
            labels_tmp = []
            for i in pred:
                labels_tmp += [int(torch.argmax(i))]
            labelsTest += labels_tmp
            if (log and batch_idx % 50 == 0):
                print("Accuracy test for epoch {} and batch {}: {}".format(epoch, batch_idx, accuracy_score(np.array(labels_tmp),yb.tolist())))
                
        predLabelsTrain = np.array(labelsTrain).flatten()       
        predLabelsTest = np.array(labelsTest).flatten()
        trueLabelsTrain = np.array(trueLabelsTrain).flatten()
        trueLabelsTest = np.array(trueLabelsTest).flatten()
        
        
        print("==> Accuracy train for epoch {}: {}".format(epoch, accuracy_score(predLabelsTrain,trueLabelsTrain)))
        print("==> Accuracy test for epoch {}: {}".format(epoch, accuracy_score(predLabelsTest,trueLabelsTest)))
    
    
    
    time_e = time.time()
    label = "batch "+str(batch_size)+", lr" + str(lr) + ", time: " + str(round((time_e - time_s), 2))
    return [train_losses, label]
```

```python
trainModel_momentum_withTest(1,.01,512, nll, initModel, .5)
```
